```python
# Importing libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (9,5)
```

```python
# Importing the data

df = pd.read_excel(r'C:\Users\PEACE\Desktop\R\IGR Correlation.xlsx')

# viewing sample data

df.head()
```

```python
# checking if there's missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# checking the data types

df.dtypes
```

```python
#CORRELATION ANALYSIS OF VARIABLES
```

```python
df.corr()
```

```python
correlation_matrix = df.corr(method="pearson")

sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix')
plt.xlabel('Features')
plt.ylabel('Features')
plt.show()
```

```python
sns.regplot(x="Avg. Area Number of Bedrooms", y="Price", data=df, scatter_kws={"color":"purple"}, line_kws={"color":"green"})
plt.title('Number of Bedrooms vs Price')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Price')
plt.show()
```

```python
sns.regplot(x="Avg. Area Income", y="Price", data=df, scatter_kws={"color":"orange"}, line_kws={"color":"violet"})
plt.title('Area Income vs Price')
plt.xlabel('Area Income')
plt.ylabel('Price')
plt.show()
```
